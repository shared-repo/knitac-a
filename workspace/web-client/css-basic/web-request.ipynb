```python
from urllib import request # 웹서버로 부터 데이터를 가져오는 모듈
```

```python
# www.daum.net 페이지를 요청해서 응답을 response에 저장

response = request.urlopen("http://www.daum.net")
```

```python
print(response.status) # 응답코드 (200 : 성공, 404 : 페이지 없음, 500 : 서버에러, ....)
```

```python
# print( response.read() ) # 응답 컨텐츠를 읽어서 binary format으로 반환
html = response.read().decode("utf-8") # 응답 컨텐츠를 읽어서 text-format 으로 반환
```

```python
# print( html )
print( len(html) )
```

```python
# BeautifulSoup 인스턴스를 만들고 html을 파싱
soup = BeautifulSoup(html, "html.parser")
```

```python
# ( css 선택자를 활용하여 ) html 요소를 찾고
numbers = soup.select("#article > div:nth-child(2) > div > div.win_result > div > div.num.win > p > span")
# 찾은 요소로부터 데이터 읽기
for number in numbers:
    print(number.text, end=" ")
```

```python
import requests

# headers : 요청 부가 정보 (여기서는 요청 클라이언트를 브라우저로 설정)
headers = { "User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36"}
naver_it_news_url = "https://news.naver.com/main/list.naver?mode=LS2D&mid=shm&sid1=105&sid2=230"
response = requests.get(naver_it_news_url, headers=headers)
response.encoding = "euc-kr"
html = response.text
response.close()
```

```python
print( html )
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "html.parser")
li_list = soup.select("div#main_content ul.type06_headline li")

# print(len(li_list))
# print(li_list[0])
```

```python
[1, 2, 3, 4, 5][-1]
```

```python
for li in li_list:
    dt_list = li.select("dt") # select : 해당되는 모든 요소 -> 리스트 반환
    # print(len(dt_list)) # 사진이 없으면 1 있으면 2
    dt = dt_list[-1] # 마지막 dt를 가져오기
    link = dt.select_one("a") # select_one : 해당되는 첫 번째 요소 -> 단일 객체 반환
    print( "제목 : ", link.text.strip() ) # strip : 문자열 양 끝에서 공백 제거
    
```
